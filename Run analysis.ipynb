```python
import pandas as pd

df = pd.read_csv('runners_data.csv')

df
```

```python
#Drop duplicates
df = df.drop_duplicates()

df = df.reset_index(drop=True)
df
```

```python
count_zero_time = (df['total_time'] == '00:00:00').sum()

print(count_zero_time)
```

```python
#Collecting runners who finished the whole race
filtered_df = df[
    (df['total_time'] != '00:00:00') & 
    (
        df['split5k'].isna() | 
        df['split10k'].isna() | 
        df['split15k'].isna() | 
        df['split20k'].isna() | 
        df['split21097k'].isna()
    )
]
filtered_df
```

```python
team_counts = df['team'].value_counts()

# Get the top 5 teams
top_5_teams = team_counts.head(5)

print(top_5_teams)

```

```python
# Filter out rows where 'total_time' is '00:00:00'
filtered_df = df[df['total_time'] != '00:00:00']

sorted_df = filtered_df.sort_values(by='total_time', ascending=False)

# Get the participant with the slowest total time
participant = sorted_df.iloc[0]

print(participant)
```
